```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.decomposition import PCA
import cv2
```

```python
img = plt.imread("C:/Users/ASUS/Downloads/abc.HEIC")
```

```python
b, g, r = cv2.split(img)
```

```python
r_scaled = r / 255
g_scaled = g / 255
b_scaled = b / 255
```

```python
com=50

pca_r = PCA(n_components=com)
pca_r_transform = pca_r.fit_transform(r_scaled)

pca_g = PCA(n_components=com)
pca_g_transform = pca_g.fit_transform(g_scaled)

pca_b = PCA(n_components=com)
pca_b_transform = pca_b.fit_transform(b_scaled)
```

```python
pca_r_original = pca_r.inverse_transform(pca_r_transform)
pca_g_original = pca_g.inverse_transform(pca_g_transform)
pca_b_original = pca_b.inverse_transform(pca_b_transform)
```

```python
img_compressed = cv2.merge((pca_b_original, pca_g_original, pca_r_original))
```

```python
plt.figure(figsize=[8, 8])
plt.imshow(img)
plt.figure(figsize=[8, 8])
plt.imshow(img_compressed)
```
